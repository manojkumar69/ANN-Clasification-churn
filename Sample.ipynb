```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle # we can use after anytime at any time of dployment
```

```python
#loading dataset
Dataset = pd.read_csv("Churn_Modelling.csv")
Dataset.head()
```

```python
## Preprosseing the data
# drop irrelevent feature like row number
Dataset = Dataset.drop(['RowNumber','CustomerId','Surname'],axis=1) #axis = colunmwise
Dataset
```

```python
#Geography, Gender categorical variable 
# Encode this 

label_encoder_gender = LabelEncoder()
Dataset['Gender']=label_encoder_gender.fit_transform(Dataset['Gender'])
Dataset
```

```python
# Geography cannot use label encoder . boz more category is there. it should be 0/1....USe OHE
#OHE -Geography

from sklearn.preprocessing import OneHotEncoder

OneHot_encode_Geo = OneHotEncoder()

geo_enocode= OneHot_encode_Geo.fit_transform(Dataset[['Geography']])
geo_enocode
```

```python
geo_enocode.toarray()
```

```python
OneHot_encode_Geo.get_feature_names_out(['Geography'])
```

```python
geo_encode_df= pd.DataFrame(geo_enocode.toarray(),columns=OneHot_encode_Geo.get_feature_names_out(['Geography']))
geo_encode_df
```

```python
# combine all columns with original Data

Dataset = pd.concat([Dataset.drop('Geography', axis = 1), geo_encode_df],axis=1)
Dataset.head()
```

```python
##Save the encoder and scalers

with open('labet_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encode_geo.pkl','wb') as file:
    pickle.dump(OneHot_encode_Geo,file)

```

```python
Dataset.head()
```

```python
# divide data set to independant and dependant features
#Exited -dependent feature

x= Dataset.drop('Exited', axis=1)
y= Dataset['Exited']

# Split the data in training and testing sets
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)

#scale the features

scaler =StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
X_train
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
###########ANN Implementation####################

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
(X_train.shape[1],)
```

```python
# Building ANN model
#64 (anything) neurons relu is one of my activation fn. we can use anything
model=Sequential([
    Dense(64,activation='relu',input_shape= (X_train.shape[1],)), ##HL1 connected with input layer
    Dense(32, activation='relu'),# HL2 
    Dense(1, activation='sigmoid') # Output layer
]


)
```

```python
model.summary()
```

```python
#oneway
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01) 
loss=tensorflow.keras.losses.BinaryCrossentropy()
#adam is one of optimizers
loss
```

```python
# To do Forword propegation and Backwrd propegation. compile  the model
#binary_crossentropy - one of keras api

model.compile(optimizer=opt, loss="binary_crossentropy", metrics=['accuracy'])
```

```python
#setup tensorboard
#tensorboard to visualize the all logs while training model
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") #time format to take logs
#when we train our moel , it ill create logs in this dir

tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## setup Early stopping
early_stop_callback =  EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True) # till 5 ephocs its ill wait
```

```python
## ************TRAINING THE MODEL *************************************

history = model.fit(
    X_train, y_train,validation_data=(X_test,y_test),epochs = 100,
    callbacks = [tensorflow_callback,early_stop_callback]
)
```
